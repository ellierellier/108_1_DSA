```python
sequence = [1, 3, 2, 1]
strict = 0
prev = 0
cur = 1

def almostIncreasingSequence(sequence, cur, prev, strict):
    if cur < len(sequence)-1 and sequence[cur] <= sequence[prev] and strict < 1:
        temp_list = sequence.pop(cur)
        strict += 1
        almostIncreasingSequence(temp_list, cur, prev, strict)
        
    else:
        prev += 1
        cur += 1
        if cur < len(sequence)-1:
            almostIncreasingSequence(sequence, cur, prev, strict)
        return True
    return False

almostIncreasingSequence(sequence, cur, prev, strict)
```
